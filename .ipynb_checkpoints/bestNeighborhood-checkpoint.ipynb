## INTRODUCTION
In my opinion, the best neighborhood in Pittsburgh is the one with the best air quality.  According to a scale called the AQI (Air Quality Index), air is ranked from 0 - 301.  0 being the best, and 301 and above being extremely dangerous.  A lower air quality in a particular neighborhood will call for healthier residents that are more motivated to exercies and prone to being outside.  A higher air quality could lead to more people with difficulty breathing, and a less productive community due to the massive amounts of pollution in the air.  In my opinion, by having a lower air quality, the better the neighborhood.  

## THE METRIC
The metric I will be using is the AQI (Air Quality Index).  The most perfect air quality is 0, but an acceptable amount ranges from 0 - 100.  An unhealthy amount ranges from 101 - 200, and a dangerous AQI goes from 201 and above.  For this project, I am using the Allegheny County Air Quality from WPRDC.  For each neighborhood, there is a specific number that coordinates with the AQI.  

## THE BEST NEIGHBORHOOD

```python
import pandas as pd
%matplotlib inline
```

```python
aqi = pd.read_csv("aqi.csv")
aqi
```

```python
list_aqi = aqi.groupby("site").mean()
list_aqi
```

```python
sorted_aqi = list_aqi.sort_values(by=["index_value"], ascending=True)
sorted_aqi
```

```python
sorted_aqi['index_value'].plot(kind="bar", width = 0.8, figsize=(25,10), ylim =(0,50))
```
